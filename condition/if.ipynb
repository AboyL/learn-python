```python
a=1
if a==1:
    print('判断')
elif a==2:
    print('不判断')
else:
    print('不判断')
```

```python
b=3
match b:
    case 1 | 3:
        print('判断')
    case 2 | 4:
        print('不判断')
    case _:
        print('不判断2')
```

## 循环

```python
c=1
while c<=10:
    if c==5:
        break
    if c==3:
        c+=1
        continue
    print(c)
    c+=1
```

```python
list1=[1,2,3,4,5]
while list1:
    print(list1.pop())
```

for循环
更加关注遍历

```python
# 使用for遍历列表
list=[1,2,3]
for i in list:
    print(i)

# 使用for遍历字典
dict={'name':'张三','age':18}
for key,value in dict.items():
    print(key,value)
print(dict.items())

# 使用for变量枚举类型获取编号
list=[1,2,3]
for i,j in enumerate(list):
    print(i,j)
```

```python
# for的推导式
list=[1,2,3]
list1=[i*2 for i in list]
print(list1)

# 带条件的推导式
list2=[i*2 for i in list if i>1]
```

```python
# 打印九九乘法表
for i in range(1,10):
    for j in range(1,i+1):
        print(f'{j}*{i}={i*j}',end=' ')
    print()
```
