```python
print("%s is %s than %s" %("Beautiful", "better", "ugly"))
print("{1} is {2} than {0}".format("ugly", "Beautiful", "better"))
print("{B} is {b} than {u}".format(u = "ugly", B = "Beautiful", b = "better"))
u = "ugly"; B = "Beautiful"; b = "better"
print(f"{B} is {b} than {u}")
```

```python
print(f"{1000:010}")
print(f"{1000:10}")
print(f"{1000:10.2f}")
# print(f"{1000:10.2d}") # ValueError: Unknown format code 'd' for object of type 'float'
print(f"{1000:10.2e}")
print(f"{1000:10.2E}")
print(f"{1000:10.2%}")
print(f"{1000:10.2%}")
# 对字符串进行格式化
print(f"{'Beautiful':>10}")
print(f"{'Beautiful':<10}")
print(f"{'Beautiful':^10}")

```
