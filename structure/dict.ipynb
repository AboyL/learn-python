字典相关处理

```python
list1=['a','b','c']
list2=[1,2,3]
dict1=dict(zip(list1,list2))
print(dict1) # {'a': 1, 'b': 2, 'c': 3}
```

```python
# 字典赋值
list1 = [ 1, 2, 3]
tuple1 = ( 'abc', list1 ) # tuple1 = ( 'abc', [1, 2, 3] ) 不可以作为字典的key
tuple1 = ( 'abc', 'list' ) # tuple1 = ( 'abc', 'list' ) 可以作为字典的key
print( type( tuple1 ))
dict1 = { tuple1: 1 }
my_dict = {('a', 1): 'apple', ('b', 2): 'banana'} # 可以作为字典的key
```

```python
# 引用问题? 跟js的一致
dict1={'a':1}
dict2=dict1
dict2['a']=2
print(dict1) # {'a': 2}
print(dict2) # {'a': 2}
dict3=dict1.copy()
dict3['a']=3
print(dict1) # {'a': 2}
print(dict3) # {'a': 3}
dict4={'a':dict1}
dict4['a']['a']=4
print(dict1) # {'a': 4}
print(dict4) # {'a': {'a': 4}}
```
