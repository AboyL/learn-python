综合内容

```python
list_demo = ['a', 'b', 'c', 'd', 'e', ['f','g']]
print(type(list_demo))
print(list_demo)
list_demo.insert(0,"first")
print(list_demo)
list_demo.insert(-1,"last")
print(list_demo)
list_demo.append("last_one")
print(list_demo)
list_demo.extend("last_one")
print(list_demo)
removeed=list_demo.remove("last_one") # 删除元素
print(list_demo)
print(removeed) # None 没有值
result = list_demo.pop(0) # 弹出元素并且返回
print(f"list is {list_demo}")
print(f"result is {result}")
```
